```python
class No:
    def __init__(self, valor):
        self.valor = valor
        self.esquerda = None
        self.direita = None
class ArvoreBinaria:
    def __init__(self):
        self.raiz = None
    def incluir_no(self, valor):
        self.raiz = self._incluir_no(self.raiz, valor)
    def _incluir_no(self, raiz, valor):
        if raiz is None:
            return No(valor)
        if valor < raiz.valor:
            raiz.esquerda = self._incluir_no(raiz.esquerda, valor)
        elif valor > raiz.valor:
            raiz.direita = self._incluir_no(raiz.direita, valor)
        return raiz
    def listar_arvore(self):
        return self._listar_arvore(self.raiz)
    def _listar_arvore(self, raiz):
        return (
            self._listar_arvore(raiz.esquerda) +
            [raiz.valor] +
            self._listar_arvore(raiz.direita)
        ) if raiz else []
    def pesquisar_no(self, valor):
        return self._pesquisar_no(self.raiz, valor)

    def _pesquisar_no(self, raiz, valor):
        return raiz is not None and (raiz.valor == valor or
          (valor < raiz.valor and self._pesquisar_no(raiz.esquerda, valor)) or
             (valor > raiz.valor and self._pesquisar_no(raiz.direita, valor)))
    def excluir_no(self, valor):
        self.raiz = self._excluir_no(self.raiz, valor)
    def _excluir_no(self, raiz, valor):
        if raiz is None:
            return raiz
        if valor < raiz.valor:
            raiz.esquerda = self._excluir_no(raiz.esquerda, valor)
        elif valor > raiz.valor:
            raiz.direita = self._excluir_no(raiz.direita, valor)
        else:
            if raiz.esquerda is None:
                return raiz.direita
            elif raiz.direita is None:
                return raiz.esquerda
            raiz.valor = self._encontrar_sucessor(raiz.direita)
            raiz.direita = self._excluir_no(raiz.direita, raiz.valor)
        return raiz
    def _encontrar_sucessor(raiz):
        while raiz.esquerda is not None:
            raiz = raiz.esquerda
        return raiz.valor
def menu():
    arvore = ArvoreBinaria()
    while True:
        print("\nMenu da árvore Binaria:")
        print("1. Incluir Nó")
        print("2. Listar Árvore")
        print("3. Pesquisar Nó")
        print("4. Excluir Nó")
        print("5. Sair")
        escolha = input("Escolha entre (1-5): ")
        if escolha == "1":
            valor = int(input("informe o valor do nó a ser incluído: "))
            arvore.incluir_no(valor)
            print(f"Nó {valor} incluído na árvore.")
        elif escolha == "2":
            print("Árvore:", arvore.listar_arvore())
        elif escolha == "3":
            valor = int(input("informe o valor do nó a ser encontrado: "))
            if arvore.pesquisar_no(valor):
                print(f"O valor {valor} foi encontrado na árvore.")
            else:
                print(f"O valor {valor} não foi encontrado na árvore.")
        elif escolha == "4":
            valor = int(input("informe o valor do nó a ser excluído: "))
            arvore.excluir_no(valor)
            print(f"Nó {valor} foi excluído da árvore.")
        elif escolha == "5":
            print("programa encerrado. Até logo!!")
            break
        else:
            print("Opção não encontrada. Tente novamente.")
if __name__ == "__main__":
    menu()
```
